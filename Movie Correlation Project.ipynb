```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the figure sizes of the plots created
import numpy as np
```

```python
# Reading the data
df = pd.read_csv("movies.csv")
```

```python
# Looking the 5 first rows of the data
df.head()
```

```python
# Verifying how many columns in rows are in this data set
df.shape
```

```python
# Looping to verify if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Seeing data types
df.dtypes
```

```python
# changing data type and dropping the NA values in these columns
df.dropna(subset = ['budget'], inplace = True)
df.dropna(subset = ['gross'], inplace = True)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Create a new year column
df.drop('year_correct', axis = 1)
```

```python
# Ordering by the movies that has the most gross
df = df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Looking at the correlations between the columns that only has numbers
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)

plt.title('Budget Correlation', fontsize = 18)
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
# Gross is highly correlated with the budget (positive correlation)
```

```python
# Converting companies names to numbers, to include in the heatmap above
df_number = df

for col_name in df_number.columns:
    if(df_number[col_name].dtype == 'object'):
        df_number[col_name] = df_number[col_name].astype('category')
        df_number[col_name] = df_number[col_name].cat.codes

        
df_number
```

```python
# Plotting the heatmap again 

correlation_all_fields = df_number.corr()
sns.heatmap(correlation_all_fields, annot = True)

plt.title('Correlation for all Fields', fontsize = 18)
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
```

```python
corr_pairs = correlation_all_fields.unstack()
sorted_pairs = corr_pairs.sort_values(ascending = False)
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr = high_corr[(high_corr) < 1]
high_corr

# We discovered that gross and budget are high correlated and votes are also correlated with gross
```
